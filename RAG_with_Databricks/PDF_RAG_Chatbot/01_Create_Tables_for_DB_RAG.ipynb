## Details of Code Below
1. Auto generate primary key in `docs_text` table in first cell which is used as the **index**.
2. Tracking row level changes to the delta table be setting `enabledChangeDataFeed = True`.

* Name of folder to insert tables: `workspace.llm_rag_demos`
* This is not the same as the volume.

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.llm_rag_demos.docs_text (id BIGINT GENERATED BY DEFAULT AS IDENTITY, text STRING) tblproperties (delta.enableChangeDataFeed = true);
```

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.llm_rag_demos.docs_track (file_name STRING) tblproperties (delta.enableChangeDataFeed = true);
```
